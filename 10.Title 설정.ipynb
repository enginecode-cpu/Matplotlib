### 타이틀 설정
- title() 함수를 사용해서 그래프의 제목을 설정할 수 있다.
- 위치와 오프셋 지정하기
    - loc를 이용해서 위치를 지정.
    - pad를 이용해서 그래프와의 간격(오프셋)을 설정할 수 있다.
    - fontdict를 이용해서 font를 설정할 수 있다.
    

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph')

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, np.exp(a), '-.')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in')
plt.title('Exp graph')

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, np.exp(a), '-.')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in')
plt.title('Exp graph', loc="right")

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, np.exp(a), '-.')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in')
plt.title('Exp graph', loc="right", pad=20)

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, np.exp(a), '-.')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in')

plt.title('Exp graph', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('Exp graph', fontdict=title_font, loc='left', pad=20)

plt.show()
```
