### Matplotlib 수평 막대 그래프 그리기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./seoul.csv", encoding="cp949")
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
new_df = pd.concat([df], ignore_index=True)
new_df
```

```python
y = np.arange(31)
day = []
temp = []
for i, col in enumerate(new_df):
    # print(i, col)
    if i == 0:
        for j in new_df[col]:
            day.append(j)
    elif i == 4:
        for j in new_df[col]:
            temp.append(j)
    else:
        continue
```

```python
plt.barh(day[:31], temp[:31])
plt.title("-".join(day[0].split('-')[:2]))
plt.yticks(y, [str(int(''.join(d.split('-')[-1]))) for d in day[:31]])
plt.figure(figsize=(100, 5))
plt.show()
```
