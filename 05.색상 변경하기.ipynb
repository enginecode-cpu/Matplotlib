### 색상 변경하기
plot() 안에 color 키워드를 이용해서 색을 표현할 수 있다. hexcode를 이용할 수 있다.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color="#27ae60")
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], '*', color="red")
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```
