### Matplotlib 상자 그래프 그리기
- 상자 그래프은 가공하지 않은 자료를 그대로 이용하는 것이 아니라, 자료에서 얻어낸 최댓값, 최솟값, 상위 1/4, 2/4(중앙), 3/4에 위치한 값을 보여주는 그래프이다.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
result = []
for i in range(10):
    result.append(np.random.randint(1000))
```

```python
result.sort()
result
```

```python
plt.boxplot(result)
plt.show()
```

```python
result = np.array(result)

print(f'1/4: {np.percentile(result, 25)}')
print(f'2/4: {np.percentile(result, 50)}')
print(f'3/4: {np.percentile(result, 75)}')
```

```python
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data :
    if row[-1] != '':
        result.append(float(row[-1]))
```

```python
result.sort()
```

```python
plt.boxplot(result)
plt.show()
```

```python
result_nparray = np.array(result)

print(f'min: {np.percentile(result_nparray, 0)}')
print(f'1/4: {np.percentile(result_nparray, 25)}')
print(f'2/4: {np.percentile(result_nparray, 50)}')
print(f'3/4: {np.percentile(result_nparray, 75)}')
print(f'max: {np.percentile(result_nparray, 100)}')
```
