### 그리드 설정하기
- grid() 함수를 이용한다. 인자로는 boolean 값을 넣어야한다.
- grid() 함수 내부에 축을 설정할 수 있다.
    - ex. grid(True, axis='y') : y축만 선을 그린다.
- 축의 style을 지정할 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y')

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='steelblue', alpha=0.5, linestyle='-.')

plt.show()
```
