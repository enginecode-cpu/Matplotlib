### 여러 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'r--', a, a**2, 'bo', a, a**3, 'g-.')
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
function1 = np.exp(x)

plt.plot(x, x, 'r-', x, function1, 'b-.')
plt.show()
```
