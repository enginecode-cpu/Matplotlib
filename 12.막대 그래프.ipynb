### Matplotlib 막대 그래프 그리기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./seoul.csv", encoding="cp949")
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
new_df = pd.concat([df], ignore_index=True)
new_df
```

```python
x = np.arange(31)
day = []
temp = []
for i, col in enumerate(new_df):
    # print(i, col)
    if i == 0:
        for j in new_df[col]:
            day.append(j)
    elif i == 4:
        for j in new_df[col]:
            temp.append(j)
    else:
        continue
```

```python
plt.bar(day[:31], temp[:31])
plt.title("-".join(day[0].split('-')[:2]))
plt.xticks(x, [str(int(''.join(d.split('-')[-1]))) for d in day[:31]])
plt.figure(figsize=(100, 5))
plt.show()
```

### bar style

속성|설명|default
:---|:---|:---
width|막대의 너비|`0.8`
align|tick과 막대의 위치를 조절. ‘edge’로 설정하면 막대의 왼쪽 끝에 x_tick이 표시|`center`
color|막대의 색을 지정
edgecolor|막대의 테두리 색을 지정
linewidth| 테두리의 두께를 지정
tick_label| array 형태로 지정하면, tick에 array의 문자열을 순서대로 나타낼 수 있습니다.
log=True| 설정하면, y축이 로그 스케일로 표시
